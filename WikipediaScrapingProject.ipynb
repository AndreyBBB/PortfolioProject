```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
#pulling all the data from the web-page to our 'soup'
```

```python
print(soup.prettify())
# can print the whole pulled HTML in a slightly more readable manner
```

```python
table = soup.find_all('table')[0]
# supposed to be pulling only the 1st table from the soup
# it didn't work properly, the issue was that in the video the table has index [1], 
# while after the update of the web-page my table has index [0]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
print(world_titles)
# it still doesn't work as in the video, because the table has different structure 
# and by pulling all the 'th' i got also all the row headers (numbers 1-50)
# so I still have to filter this headers manually
```

```python
indexes_to_copy = [1,2,3,4,5,6,7,8,12] # also cut off column #0 - Rank - because there are no rank info in the rows
world_titles_cut = [world_titles[i] for i in indexes_to_copy]
print (world_titles_cut)
# I have just manually cut off all the unneeded headers
```

```python
world_table_titles = [title.text.strip() for title in world_titles_cut]
# it's a list comprehension procedure, and the '.strip' cuts off '\n' part of every header
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
print(column_data)
```

```python
df = df.drop(df.index) # deleting all the rows from our data frame to avoid duplicating rows each time we run this chunk of code
for row in column_data[2:]: # using the index we are cutting off 2 first rows, which are empty (probable these are the headers?)
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
#    print(individual_row_data)
# 'print' is in the loop, because in the moment 'individual_row_data' has only 1 row of information
# now we need to put all this in our dataframe
# the only problem I see - the original table has an icon to show if the company is state-owned or not 
# and in that case we pull just an empty field
    length = len(df)
    df.loc[length] = individual_row_data
# another issue - in my case the ranks are not in the rows data, they are headers, si I filtered them out on a previous step...
```

```python
df
```

```python
# Now we got the whole table, but don't have Ranks! 
# Because Ranks now are somehow structured as column headers and were filtered out by me pulling headers to out columns
# So now the whole column Rank is deleted. It can be a serious problem, 
# because we can sort the table on the web-page before scraping and can absolutely lose the hierarchy of ranks
# in theory we can restore ranks, using the logic of ranking and manually sorting the rows, but in general it's really bad practice
```
